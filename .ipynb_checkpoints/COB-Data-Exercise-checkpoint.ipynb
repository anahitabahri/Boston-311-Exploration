```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")

import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
data = pd.read_csv('../311__Service_Requests.csv')
```

```python
data['open_date'] = data['OPEN_DT'].str[:10]
data['open_date'] = pd.to_datetime(data['open_date'], format="%m/%d/%Y")
```

```python
data['year'] = data['OPEN_DT'].str[6:10].astype(int)
data['month'] = data['OPEN_DT'].str[:2].astype(int)
```

## Only care about data after Jan 2016! 

```python
data = data[(data.year == 2016) | (data.year == 2017)]
```

```python
data.to_csv('data.csv',index=False)
```

## Read in updated data

```python
data = pd.read_csv('data.csv')
```

### Explore histograms, countplots

```python
sns.countplot(y='CASE_STATUS', color='#7C7CE0', data=data)
plt.xlabel('Count')
plt.ylabel('Case Status')
plt.title('Count of Requests by Status')
```

```python
sns.countplot(y='OnTime_Status', color='#7C7CE0', data=data)
plt.xlabel('Count')
plt.ylabel('On Time Status')
plt.title('Count of Requests by On Time Status')
```

```python
sns.countplot(y='SUBJECT', color='#7C7CE0', data=data)
plt.xlabel('Count')
plt.ylabel('Subject')
plt.title('Count of Requests by Subject')
```

## How about Jan 2016 vs. Jan 2017?

```python
jan_data = data[(data.month == 1)]
```

```python
sns.countplot(y='SUBJECT', hue="year", data=jan_data)
plt.xlabel('Count')
plt.ylabel('Subject')
plt.title('Count of Requests by Subject')
```

```python
sns.countplot(y='neighborhood', hue="year", data=jan_data)
plt.xlabel('Count')
plt.ylabel('Neighborhood')
plt.title('Count of Requests by Neighborhood')
```

```python
sns.countplot(y='Source', hue="year", data=jan_data)
plt.xlabel('Count')
plt.ylabel('Source')
plt.title('Count of Requests by Source')
```

```python
sns.countplot(x="year", hue="OnTime_Status", data=jan_data) 
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Count of Requests: 2016 vs. 2017')
```

### Let's look into transportation - traffic division

```python
transport = jan_data[(jan_data.SUBJECT == 'Transportation - Traffic Division')]
```

```python
sns.countplot(y='REASON', hue="year", data=transport)
plt.xlabel('Count')
plt.ylabel('Reason')
plt.title('Count of Requests by Reason')
```

```python
transport = transport[(transport.REASON == 'Enforcement & Abandoned Vehicles') | (transport.REASON == 'Signs & Signals')]
```

```python
sns.countplot(y='TYPE', hue="year", data=transport)
plt.xlabel('Count')
plt.ylabel('Type')
plt.title('Count of Requests by Type')
```
